```python
import numpy as np


class BayesianLinearRegression:
    
    def __init__(self,X,Y,a,b):
        
        self.x=X
        self.y=Y
        self.a=a
        self.b=b
        self.jitter=1e-8
        
    def MLE(self):
        xTx=np.matmul(self.x.T,self.x)
        xTx_inv=np.linalg.inv(xTx+self.jitter)
        xTy = np.matmul(self.x.T, self.y)
        w_MLE=np.matmul(xTx_inv,xTy)
        
        self.w_MLE=w_MLE
        
        return w_MLE
    
    def MAP(self):
        Lambda_inv=np.linalg.inv(np.matmul(self.a*self.x.T,self.x)+self.b*np.eye(self.x.shape[1]))
        xTy=self.a*np.matmul(self.x.T,self.y)
        w_MAP=np.matmul(Lambda_inv,xTy)

        self.w_MAP=w_MAP
        self.Lambda_inv=Lambda_inv
        return w_MAP, Lambda_inv
    
    def predictive_distribution(self, X_star):
        mean_star = np.matmul(X_star, self.w_MAP)
        var_star = 1.0/self.a + np.matmul(X_star, np.matmul(self.Lambda_inv, X_star.T))
        return mean_star, var_star
```

```python
class xdes:
    
    def __init__(self,X):
        self.x=X
    
    
    def iden(self):
        return self.x
    
    def mon(self,N):
        x_mon=np.ones(self.x.shape[0])[:,None]
        for s in range(1,N+1):
            x_mon=np.hstack((x_mon,self.x**s))
        
        return x_mon
    
    def four(self,N):
        xfour=np.hstack((np.zeros(self.x.shape[0])[:,None],np.ones(self.x.shape[0])[:,None]))
        for s in range(1,N+1):
            xfour=np.hstack((xfour,np.hstack((np.sin(s*np.pi*self.x),np.cos(s*np.pi*self.x)))))
        
        return xfour
    
    def lege(self,N):
        w=sc.special.legendre(1)
        lege_2=w(self.x)
        lege_1=lege_2
        for s in range (2,N+1):
            w=sc.special.legendre(s)
            lege_2=w(self.x)
            lege_1=np.hstack((lege_1,lege_2))
        return lege_1
```

```python
from pyDOE import lhs
from scipy.stats import norm
import scipy as sc
```

```python
if __name__ == "__main__":
    M=8
    N=500
    a=5
    c=0.1
    #create random input and output data
    #x=np.linspace(0,2,N)
    x=2*lhs(1,N)
    #x=np.reshape(x,(N,1))
    y=np.exp(x)*np.sin(2*np.pi*x)+np.random.normal(0,0.7071,N)[:,None]
    
    #basis selection
    b=xdes(x)
    x_four=b.four(M)
    x_iden=b.iden()
    x_lege=b.lege(M)
    x_mon=b.mon(M)
    
    #define model
    m_four = BayesianLinearRegression(x_four,y,a,c)
    m_iden = BayesianLinearRegression(x_iden,y,a,c)
    m_lege = BayesianLinearRegression(x_lege,y,a,c)
    m_mon = BayesianLinearRegression(x_mon,y,a,c)
    
    #MLE MAP estimates
    w_MLE_four=m_four.MLE()
    w_MLE_iden=m_iden.MLE()
    w_MLE_lege=m_lege.MLE()
    w_MLE_mon=m_mon.MLE()
    
    
    w_MAP_four,Lambda_inv_four=m_four.MAP()
    w_MAP_iden,Lambda_inv_iden=m_iden.MAP()
    w_MAP_lege,Lambda_inv_lege=m_lege.MAP()
    w_MAP_mon,Lambda_inv_mon=m_mon.MAP()

    x_star=np.linspace(0,2,N)[:,None]
    x_star_graph=x_star
    
    b=xdes(x_star)
    x_star_four=b.four(M)
    x_star_iden=b.iden()
    x_star_lege=b.lege(M)
    x_star_mon=b.mon(M)
   
    y_MLE_four=np.matmul(x_star_four,w_MLE_four)
    y_MAP_four=np.matmul(x_star_four,w_MAP_four)
    
    y_MLE_iden=np.matmul(x_star_iden,w_MLE_iden)
    y_MAP_iden=np.matmul(x_star_iden,w_MAP_iden)
    
    y_MLE_lege=np.matmul(x_star_lege,w_MLE_lege)
    y_MAP_lege=np.matmul(x_star_lege,w_MAP_lege)
    
    y_MLE_mon=np.matmul(x_star_mon,w_MLE_mon)
    y_MAP_mon=np.matmul(x_star_mon,w_MAP_mon)
    
    #extracting 500 samples from predictive posterior distribution
    num_samples = 500
    mean_star_four, var_star_four = m_four.predictive_distribution(x_star_four)
    samples_four = np.random.multivariate_normal(mean_star_four.flatten(), var_star_four, num_samples)
    
    mean_star_iden, var_star_iden = m_iden.predictive_distribution(x_star_iden)
    samples_iden = np.random.multivariate_normal(mean_star_iden.flatten(), var_star_iden, num_samples)
    
    mean_star_lege, var_star_lege = m_lege.predictive_distribution(x_star_lege)
    samples_lege = np.random.multivariate_normal(mean_star_lege.flatten(), var_star_lege, num_samples)
    
    mean_star_mon, var_star_mon = m_mon.predictive_distribution(x_star_mon)
    samples_mon = np.random.multivariate_normal(mean_star_mon.flatten(), var_star_mon, num_samples)
        
```
